```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("carrie1/ecommerce-data")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import os

# Assuming the downloaded dataset contains a CSV file, let's find it.
# For 'carrie1/ecommerce-data', a common filename is 'data.csv' or similar.
# Let's list the files in the downloaded path to be sure.

print("Files in the downloaded dataset directory:")
for root, dirs, files in os.walk(path):
    for file in files:
        print(os.path.join(root, file))

# Assuming 'data.csv' is the main file to load
# If the actual file name is different, you will need to update 'data.csv' below
csv_file_path = os.path.join(path, 'data.csv') # Common filename for this dataset, adjust if needed

df = pd.read_csv(csv_file_path, encoding = 'ISO-8859-1')
```

```python
sales=df.groupby("Description")["Quantity"].sum().head(10)
```

```python
sales.plot(kind="bar",figsize=(10,5))
plt.xlabel("Product")
plt.ylabel("Quantity")
plt.show()
```

```python
sales.plot(kind="line",marker="o",figsize=(10,5))
plt.xlabel("Product")
plt.ylabel("Quantity")
plt.show()
```
